```python
temp = df_biometric_1["district_clean"].copy()

# create a super-normalized version to detect near duplicates
df_biometric_1["district_norm"] = (temp
                                  .str.lower()
                                  .str.replace(r"[^a-z0-9]", "", regex=True))  # remove space, hyphen, symbols

# find groups where multiple names map to same normalized form
duplicates = (df_biometric_1.groupby("district_norm")["district_clean"]
              .nunique()
              .sort_values(ascending=False))

print(duplicates[duplicates > 1].head(50))
```

```python
dup_keys = duplicates[duplicates > 1].head(20).index

for key in dup_keys:
    names = df_biometric_1.loc[df_biometric_1["district_norm"] == key, "district_clean"].unique()
    print(key, "->", names)
```

```python
for key in dup_keys:
    subset = df_biometric_1[df_biometric_1["district_norm"] == key]
    common = subset["district_clean"].value_counts().idxmax()
    df_biometric_1.loc[df_biometric_1["district_norm"] == key, "district_clean"] = common
```
